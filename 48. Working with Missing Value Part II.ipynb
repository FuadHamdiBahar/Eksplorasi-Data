```python
import pandas as pd
import numpy as np

dict = {'First':[100, 90, 95, 100, np.nan, 100],
        'Second': [30, 45, np.nan, 50, np.nan, 40],
        'Third':[np.nan, 40, 98, 99, np.nan, 90]}

df = pd.DataFrame(dict)
df
```

```python
# mengganti yang spesifik nilainya
df.replace(to_replace=np.nan, value=1)
```

```python
# mengganti list of value
df.replace(to_replace=[np.nan, 100], value=1)
```

```python
# bfill ffill dan pad
df.replace(to_replace=np.nan, method='pad')
```

```python
# linear, time, index, pad, other
df.interpolate(method='nearest')
```
