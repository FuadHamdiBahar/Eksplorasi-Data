```python
import pandas as pd
import numpy as np

dict = {'First':[100, 90, 95, 100, np.nan, 100],
        'Second': [30, 45, np.nan, 50, np.nan, 40],
        'Third':[np.nan, 40, 98, 99, np.nan, 90]}

df = pd.DataFrame(dict)
df
```

```python
df.dropna(axis=0, how='any')
```

```python
df.dropna(axis=0, how='all')
```

```python
df.fillna(value=0)
```

```python
df.sort_values(by=['First'], inplace=True)
df
```

```python
# method ffill mengambil data sebelumnya untuk disimpan data saat ini
# n[kol] = n[kol][i - 1]
# method bfil itu kebalikannya
df.fillna(method='ffill')
```

```python
df.fillna(method='bfill')
```
