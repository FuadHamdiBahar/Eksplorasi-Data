```python
import pandas as pd

df = pd.read_csv('./Data/Placement.csv').drop(columns='placed')
df.head()
```

```python
# xi - mean / standar_deviasi
(df['cgpa'][0] - df['cgpa'].mean()) / df['cgpa'].std()
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
scaled = ss.fit_transform(df)
new_df = pd.DataFrame(data=scaled, columns=df.columns)
new_df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(16, 5))

plt.subplot(1,2,1)
sns.histplot(new_df['cgpa'], kde=True)

plt.subplot(1,2,2)
sns.histplot(new_df['placement_exam_marks'], kde=True)

plt.show()
```

```python
import numpy as np

std = np.std(new_df['cgpa'])
mean = round(np.mean(new_df['cgpa']), 2)
print("Standar Deviasi: {} \nRata-Rata: {}".format(std, mean))
```

```python
std = np.std(new_df['placement_exam_marks'])
mean = round(np.mean(new_df['placement_exam_marks']), 2)
print("Standar Deviasi: {} \nRata-Rata: {}".format(std, mean))
```
