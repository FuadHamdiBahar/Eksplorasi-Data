```python
import matplotlib.pyplot as plt

bulan = ['Jan', 'Feb', 'Mar', 'Apr', 'Mei', 'Jun']
pakaian = [30, 40, 50 ,60, 80, 100]
pokok = [50, 47, 52, 55, 53, 50]
teknologi = [100, 30, 30, 40, 30, 40]
```

```python
plt.plot(bulan, pakaian, marker='o')
plt.plot(bulan, pokok, marker='s')
plt.plot(bulan, teknologi, marker='^')
plt.legend(['Pakaian', 'Pokok', 'Teknologi'])
plt.title("Tingkat Pengeluaran Masyarakat")
plt.ylabel("Juta Rupiah")
```

```python
import numpy as np 
import matplotlib.pyplot as plt 

X = ['Pakaian','Pokok','Teknologi']
Ygirls = [10,20,40]
Zboys = [25,20,25]

X_axis = np.arange(len(X))

plt.bar(X_axis - 0.2, Ygirls, 0.4, label = 'Laki-Laki')
plt.bar(X_axis + 0.2, Zboys, 0.4, label = 'Perempuan')

plt.xticks(X_axis, X)
plt.xlabel("Gender")
plt.ylabel("Juta Rupiah")
plt.title("Pengeluaran Masyarakat berdasarkan Gender")
plt.legend()
plt.show()
```
