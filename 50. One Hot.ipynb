```python
import pandas as pd

df = pd.read_csv('./Data/Universitas.csv').drop(axis=1, columns='Program Studi').drop_duplicates(subset=['Nama_PT'])
df = df.dropna().reset_index(drop=True)
df
```

```python
encoded = pd.get_dummies(df, columns=['Kota'], prefix='Kota', dtype=int)
encoded
```

```python
from sklearn.preprocessing import OneHotEncoder

# make encoder
encoder = OneHotEncoder()
encoded = encoder.fit_transform(df[['Kota']])
# mengambil nilau unique
col = df['Kota'].unique()
# menngubah array ke df
encoded_kota = pd.DataFrame(encoded.toarray(), columns=col, dtype=int)
# menggabung encoded kota dengan nama pt
pd.concat([df[['Nama_PT']], encoded_kota], axis=1)
```
