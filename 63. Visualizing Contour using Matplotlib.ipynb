```python
import numpy as np
```

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)

Z = f(X, Y)
Z.shape
```

```python
import matplotlib.pyplot as plt

plt.contour(Z)
plt.show()
```

```python
plt.contour(Z, colors='black')
plt.contour(Z, cmap='RdGy')
plt.show()
```

```python
plt.contour(Z, cmap='RdGy')
```

```python
plt.contourf(Z, levels=20, cmap='RdGy')
plt.colorbar();
```

```python
plt.imshow(Z, origin='lower', cmap='RdGy')
plt.colorbar()
plt.axis();
```

```python
contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)

plt.imshow(Z, extent=(0, 5, 0, 5), origin='lower',
           cmap='RdGy', alpha=0.5)
plt.colorbar();
```
