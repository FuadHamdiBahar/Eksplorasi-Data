```python
import pandas as pd

df = pd.read_csv("./Data/Placement.csv").drop(columns='placed')
df.sample(5)
```

```python
scaled = (df['cgpa'][0] - df['cgpa'].min()) / (df['cgpa'].max() - df['cgpa'].min())
scaled
```

```python
from sklearn.preprocessing import MinMaxScaler

mms = MinMaxScaler()
scaled = mms.fit_transform(df)
new_df = pd.DataFrame(scaled, columns=df.columns)
new_df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(16, 4))

plt.subplot(1,2,1)
sns.histplot(new_df['cgpa'], kde=True)

plt.subplot(1,2,2)
sns.histplot(new_df['placement_exam_marks'], kde=True)

plt.show()
```
