```python
import matplotlib.pyplot as plt

x = ["A", "B", "C", "D"]
y = [3, 8, 1, 10]

plt.bar(x,y)
plt.show()
```

```python
plt.barh(x,y)
plt.show()
```

```python
x = ['A', 'B', 'C', 'D']
y1 = [10, 20, 10, 30]
y2 = [20, 25, 15, 25]
 
plt.bar(x, y1)
plt.bar(x, y2, bottom=y1)
plt.show()
```

```python
import numpy as np

x = np.array([1,2,3,4,5])
y1 = [34, 56, 12, 89, 67] 
y2 = [12, 56, 78, 45, 90] 
width = 0.4

plt.bar(x - 0.2, y1, width) 
plt.bar(x + 0.2, y2, width) 
plt.show()
```
