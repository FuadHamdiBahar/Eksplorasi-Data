```python
import numpy as np
import seaborn as sns

# Data simulation
np.random.seed(1)
data = np.random.rand(5, 5)

sns.heatmap(data)
```

```python
data
```

```python
sns.heatmap(data,
            linewidths = 0.75, 
            linecolor = "white") # Default color
```

```python
sns.heatmap(data,
            annot = True,
            annot_kws = {'size': 6})
```

```python
sns.heatmap(data,
            cbar = False)
```

```python
sns.heatmap(data,
            cmap = 'Blues')
```
